```python
import pandas as pd
import numpy as np

np.random.seed(10)


def form_unequal_group_splits(guests_data: pd.DataFrame):
    already_taken_guests = []
    group_guests_map = dict()
    for group_name, group_size in dict({"control": 50, "target": 50}).items():
        new_df = guests_data.loc[~guests_data["customer_rk"].isin(already_taken_guests)].copy()
        group_frac_to_take = min(group_size / len(new_df), 1)

        group_guests = (
            new_df
            .groupby("strata", group_keys=False)
            .apply(lambda x, frac=group_frac_to_take: x.sample(frac=frac))
            ["customer_rk"].tolist()
        )

        abs_percent_err = round(abs((len(group_guests) - group_size) / group_size) * 100, 2)
        print(f"{group_name}: Desired size = {group_frac_to_take}, \
        resulting size = {len(group_guests)}, diff = {abs_percent_err} %")

        group_guests_map[group_name] = group_guests
        already_taken_guests.extend(group_guests)

    return group_guests_map

print(form_unequal_group_splits(df))
print(form_unequal_group_splits(df))
```

```python
DAYS = 35

assets_df = pd.DataFrame(assets).dropna()
assets_df = assets_df[assets_df.closed.apply(len) > 1]
assets_df = assets_df.explode('closed')
assets_df = assets_df.reset_index(drop=True)

assets_df['dt'] = assets_df['closed'].apply(lambda x: x.dttm if not pd.isnull(x) else x)
assets_df['closed'] = assets_df['closed'].apply(lambda x: x.value if not pd.isnull(x) else x)
assets_df = assets_df[assets_df.dt >= (DATE_TO.date() - datetime.timedelta(days=DAYS))]
```

```python
sorted(assets_df.name.unique())
```

```python
# assets_df.to_csv("etf.csv", index=False)
```

```python
unique_figi = assets_df.figi.drop_duplicates()
pbar = tqdm.tqdm(unique_figi, total=len(unique_figi), leave=False)
pbar.set_description("FIGI")
for figi in pbar:
    series_closed = assets_df[assets_df.figi == figi]['closed']
    
    assets_df.loc[assets_df.figi == figi, 'increase_rate'] = (
        series_closed / series_closed.shift()
    ).fillna(1)
    
    assets_df.loc[assets_df.figi == figi, 'total_increase'] = (
        assets_df.loc[assets_df.figi == figi, 'increase_rate']
        .cumprod()
        .tail(1)
        .reset_index(drop=True)
        [0]
    )
```

```python
tmp = assets_df.groupby(["figi", "name", "currency"], as_index=False).agg(
    std_increase_rate=('increase_rate', 'std'),
    total_increase=('total_increase', 'max'),
    count=('dt', 'count'),
)
```
